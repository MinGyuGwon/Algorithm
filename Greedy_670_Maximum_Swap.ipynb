```python
def maximumSwap(num):
  num_list = list(str(num))
  flag = False
  str_num =  str(num) # 문자열 메소드 rfind( )
  print(flag)
  
  # 한번 Swap이 진행되었으면 종료
  for idx in range(len(num_list)): 
    if flag:
      print("break")
      break 
    
    temp_max = max(map(int,num_list[idx:]))
    print(f"temp_max :  {temp_max}")
    idx_print = int(num_list[idx])
    if int(num_list[idx]) < temp_max:
      print("조건문 실행")
      swap_idx = str_num.rfind(str(temp_max))
      print(f"num : {idx_print} \nswap: {temp_max}")
      num_list[idx], num_list[swap_idx] = num_list[swap_idx], num_list[idx]
      flag = True
  max_int = "".join(num_list)
  max_int = int(max_int)
  return max_int
```

```python
# 조X빈 풀이
# 성공
def maximumSwap(num):
    # 나중에 합치기 용이하게 문자열로 변환
    digits = list(str(num)) 

    # 주어진 숫자들로 가능한 최댓값
    largest = sorted(digits,reverse=True) 

    # 왼쪽부터 한 자리씩 비교 
    for i in range(len(digits)):
        # 잘못된 자릿수 i
        if digits[i]!=largest[i]:
            # 잘못된 자릿수 i 뒤의 가장 큰 값의 인덱스 idx에 저장
            idx = len(digits)-1-digits[::-1].index(max(digits[i:]))

            # 잘못된 수와 그 뒤 가장 큰 수 바꾸기
            digits[i], digits[idx] = digits[idx], digits[i]
            break
    return int("".join(digits))
```

```python
# https://leetcode.com/problems/maximum-swap/discuss/185982/Straightforward-O(n)-python

def maximumSwap2(num):
    num = [int(x) for x in str(num)]
    max_idx = len(num) - 1
    # 바꿀 자릿수의 변수
    xi = yi = 0
    for i in range(len(num) - 1, -1, -1):
        if num[i] > num[max_idx]:
            max_idx = i
        elif num[i] < num[max_idx]:
            # xi를 갱신시키는 이유: 앞자리로 바꿔야 값이 더 커지기 때문에
            xi = i
            yi = max_idx # yi이 max_idx가 갱신될 때마다 바뀌는 것이 아니고, 확실하게 max_idx의 값보다 작은 값이 있을 때만 저장이 된다. (맨 앞에 오름차순으로 정렬되어 있을 때는 무시)
    num[xi], num[yi] = num[yi], num[xi]
    return int(''.join([str(x) for x in num]))
```

```python
# https://leetcode.com/problems/maximum-swap/discuss/107066/Python-Straightforward-with-Explanation/624541

def maximumSwap3(self, num):
    a = list(map(int, str(num)))
    # 각 값의 가장 낮은 자릿수 (오른쪽)를 사전에 저장
    last = {x: i for i,x in enumerate(a)}

    # 왼쪽부터 각 자릿수에 대하여
    for i,x in enumerate(a):
        # 해당 자릿수의 값보단 큰 값이
        for d in range(9,x,-1):
            # 존재하고
            if d in last:
                # 해당 자릿수 뒤에 있다면 -> 바꾸기
                if last[d]>i:
                    a[last[d]],a[i]=a[i],a[last[d]]
                    return int(''.join(map(str,a)))
    return num
```
