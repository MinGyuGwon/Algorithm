https://leetcode.com/problems/two-sum/

```python
def twoSum(nums, target):
  for i in range(len(nums)-1):  # i는 최대 2
    for j in range(i+1,len(nums)):   # range(3,4)
      if nums[i] + nums[j] == target:
        return [i,j]
```

```python
nums = [2,7,11,15]
target = 9

for i in range(len(nums)-1):  # i는 최대 2
  for j in range(i+1,len(nums)):   # range(3,4)
    if nums[i] + nums[j] == target:
      print([i,j])
```

```python
twoSum( nums = [2,7,11,15], target = 9)  # Expected : [0,1]
```

```python
twoSum( nums = [3,2,4], target = 6)  # Expected : [1,2]
```

```python
twoSum( nums = [3,3], target = 6)  # [0,1]
```
